```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'id': range(7),
    'Title': ['Toy Story', 'A Bug\'s Life', 'Toy Story 2', 'Monsters, Inc.', 'Finding Nemo', 'Inside Out', 'The Good Dinosaur'],
    'Director': ['John Lasseter', 'John Lasseter', 'John Lasseter', 'Pete Docter', 'Andrew Stanton', 'Pete Docter', 'Peter Sohn'],
    'Year': [1995, 1998, 1999, 2001, 2003, 2015, 2015],
    'Length_minutes': [81, 95, 93, 92, 107, 95, 97]
    })
```

```python
df2 = pd.DataFrame({
    'movie_id': [0, 0, 1, 2, 4, 4, 4, 5, 6, 8],
    'critic_rating': [8.3, 7.2, 7.9, 8.1, 8.2, 8.9, 9.1, 8.3, 8.9, 9.0]
})
```

```python
df3 = pd.DataFrame({
    'movie_id': range(4),
    'Domestic_sales_millions': [191, 162, 245, 289],
    'International_sales_millions': [170, 200, 239, 272]
})
```

Create a table that shows for each movie, its movie title, year and international sales. For the movies where international sales is not available, show null instead (1 point).

```python
df4 = pd.merge(df1,
    df3,
    left_on='id',
    right_on='movie_id',
    how='outer')
```

```python
df4['International_sales_millions'] = df4['International_sales_millions'].replace(np.NaN, 'null')
df4
```

Create a table showing for each movie, its title, its critic ratings (show null if not available) and its domestic sales. Don't include movies for which domestic sales is not available (3 points).

```python
critic_sales = pd.merge(df2,
    df3,
    how='left'    
    )
critic_sales
```

```python
df5 = pd.merge(df1, 
    critic_sales,
    left_on='id',
    right_on='movie_id',
    how='left'
    )
```

```python
df5 = df5[df5['Domestic_sales_millions'].notna()]
df5
```

Create a new dataframe from df1 that sets 'Director' and 'Year' to be the joint index (1 point).

```python
joint_index = df1.set_index(['Director', 'Year'])
joint_index
```

Using the dataframe created in the last step, calculate for each director, the average length (in minutes) of their movies (2 points).

```python
joint_index.groupby('Director').mean()
```

Using the dataframe created in Step 3, print only the movies directed by John Lasseter or Pete Docter. Show only the columns for movie id and its length (1 point).

```python
joint_index.loc[['John Lasseter', 'Pete Docter']].drop('Title',axis=1)
```

Using the dataframe created in Step 3, print only the movies directed in the years 1995 or 2015. Show only the columns for movie id and its title (2 points).

```python
condition = joint_index.index.get_level_values('Year').isin([1995,2015])
joint_index[condition].drop('Length_minutes',axis=1)
```
