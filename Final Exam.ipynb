(d) Use pd.cut to create a new column 'Region' that satisfies the following conditions (3 points):

    Rows with a latitude lower than 42.4 should have Southern in the Region column
    Rows with a latitude between 42.4 and 42.8 should have Central in the Region column
    Rows with a latitude higher than 42.8 should have Northern in the Region column

```python
canada_fish['Region'] = pd.cut(canada_fish['Start Latitude'], bins=[-float("inf"), 42.4, 42.8, float("inf")], labels=['Southern', 'Central', 'Northern'])
```

(e) Make lineplots showing maximum air temperature by month and Region. You can have the month on the horizontal axis, the air temperature on the vertical axis, and different colors for different regions (5 points).
