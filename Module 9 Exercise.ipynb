```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```


Create a histogram of sepal length with the following: (3 points)
    Title the plot as 'Sepal length'
    Cal the legend 'sepal length' and show it at the upper left location
    X label name = Bins
    Y label name = Centimeters
    Bars of the color #2ab0fc
    Bars' edges of the color magenta
    Line width should be 3.5


```python
data = pd.read_csv('iris.data')
```

```python
sepal_length = data['sepal length']
```

```python
fig = plt.figure()
plt.hist(sepal_length, color='#2ab0fc', edgecolor='m', linewidth=3.5, label='sepal length')

plt.legend(loc='upper left')
plt.title('Sepal length')
plt.xlabel('Bins')
plt.ylabel('Centimeters')
```

There are three species of flowers in the dataset. The sepal length-width combination of one of the species is quite distinct from that of the other two species. Make an appropriate visualization plot to find out the distinct species. (3 points)

```python
data['sepal combo'] = data['sepal length'] * data['sepal width']
```

```python
setosa = data[data['species'] == 'Iris-setosa']
versicolor = data[data['species'] == 'Iris-versicolor']
virginica = data[data['species'] == 'Iris-virginica']
```

```python
plt.plot(setosa['sepal combo'], color='g', label='setosa')
plt.plot(versicolor['sepal combo'], color='y', label='versicolor')
plt.plot(virginica['sepal combo'], color='m', label='virginica')

plt.legend()
```

Make a plot that shows you the distribution of petal length for each of the three species. It should show median, 25th and 75th percentile values and outliers (2 points). Based on the plot, answer the following questions:
    Which species does not have any outlying values of petal length? (1 point)
    Which species always has (or have) a larger petal length than setosa? (1 point)

```python
plt.boxplot(setosa['petal length'], positions=[1])
plt.boxplot(versicolor['petal length'], positions=[2])
plt.boxplot(virginica['petal length'], positions=[3])

plt.xticks([1,2,3], ['setosa', 'versicolor', 'virginica'])
plt.title('Petal length')
plt.grid()
plt.show()
```

There are no outliers for the virginica species.

Both versicolor and virginica both have a larger petal length than setosa.
